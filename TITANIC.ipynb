## 수치형 데이터(Numeric data) to 범주형 데이터(Categorical data) 
##### 타이타닉의 나이 데이터는 0.42살 부터 80살까지로 구성되어있음
##### 연령을 20살 단위로 4개의 클래스로 만듦
##### 0 ~ 20살: 0  
##### 21 ~ 40살: 1
##### 41 ~ 60살: 2
##### 61 ~ 80살: 3

```python
age = df.Age.unique()
age = np.sort(age)
age
```

```python
df
```

```python
# 연령대별로 구별하기 위해 2차원 리스트 생성
age_list =[[0, 20], [20, 40], [40, 60], [60, 80]]

for idx, age in enumerate(age_list):
    print(age, idx)
    df.loc[(age[0] < df['Age']) & (df['Age'] <= age[1]), 'Age'] = idx

df
```

```python
survived_sex = df.groupby(['Sex', 'Survived'])
survived_sex_count = survived_sex['Survived'].count()
survived_sex_count
```

```python
survived_male = df.groupby(['Sex', 'Survived'])['Survived'].count()['male'][1]
survived_female = df.groupby(['Sex', 'Survived'])['Survived'].count()['female'][1]
```

```python
survived_age = df.groupby(['Age', 'Survived'])
survived_age_count = survived_age['Survived'].count()
survived_age_count
```

survived_age_0 = df.groupby('Age', )

```python
# 연령대별 생존자 수를 그룹화하여 저장
survived_by_age = df.groupby(['Age', 'Survived'])['Survived'].count()

# 나이 그룹 0.0에서 생존자 수 확인
survived_age_0 = survived_by_age[0.0, 1] if (0.0, 1) in survived_by_age.index else 0
survived_age_1 = survived_by_age[1.0, 1] if (1.0, 1) in survived_by_age.index else 0
survived_age_2 = survived_by_age[2.0, 1] if (2.0, 1) in survived_by_age.index else 0
survived_age_3 = survived_by_age[3.0, 1] if (3.0, 1) in survived_by_age.index else 0

print(survived_age_0, survived_age_1, survived_age_2, survived_age_3)
```

```python
# Pclass별 생존자 수
survived_pclass = df.groupby(['Pclass','Survived'])
survived_plcass_count = survived_pclass['Survived'].count()
survived_plcass_count
```

```python
# 전체 인원 수
total_count = len(df)

# Pclass별 생존자 수
survived_pclass = df[df['Survived'] == 1].groupby('Pclass')['Survived'].count()

# Pclass별 생존 비율 계산: 전체 인원 수 / Pclass별 생존자 수
survival_pclass_rate = survived_pclass / total_count

survival_rate_1 = survival_pclass_rate[1]
survival_rate_2 = survival_pclass_rate[2]
survival_rate_3 = survival_pclass_rate[3]
# 결과 출력
print(survival_pclass_rate)
```

## 성별에 따른 생존자 시각화

```python
fig, ax = plt.subplots()
sex = df.Sex.unique().tolist()
counts = [survived_male, survived_female]
bar_colors = ['tab:blue', 'tab:red']

ax.bar(sex, counts, color=bar_colors)
ax.set_title('Number of survived per sex')
```

## 나이에 따른 생존자 시각화

```python
fig, ax = plt.subplots()
age = df.Age.unique().tolist()
x = ['0.0','1.0','2.0','3.0']
counts = [survived_age_0, survived_age_1, survived_age_2, survived_age_3]
bar_colors = ['tab:blue', 'tab:red','tab:orange', 'tab:purple']
ax.bar(x, counts, color=bar_colors)
ax.set_title('Number of survived age')
```

## Pclass에 따른 생존자 시각화

```python
fig, ax = plt.subplots()
pclass = df.Pclass.unique().tolist()
counts = [survival_rate_1, survival_rate_2, survival_rate_3]
bar_colors = ['tab:red','tab:blue','tab:gray']
ax.bar(pclass,counts, color=bar_colors)
ax.set_title('Number of survived pclass')
```
